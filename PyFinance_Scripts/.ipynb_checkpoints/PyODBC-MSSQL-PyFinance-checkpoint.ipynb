```python
import pandas as pd
import yfinance as yf
from get_histdata import GetHistData_yf
from getTickers import nifty200tickers,nifty50tickers
from datetime import datetime,timedelta
import warnings
warnings.simplefilter(action='ignore', category=Warning)
```

```python
tickers=["{}.NS".format(tick) for tick in nifty50tickers()]
```

```python
df= GetHistData_yf(tickers,datetime.today()-timedelta(365))
df['Close']=df['Adj Close']
df.drop(columns='Adj Close',inplace=True)
```

```python
df
```

```python
pct_check=5
multiplier=((100-pct_check)/100)
for tick in df.Symbol.unique():
    
    df1=df[df['Symbol']==tick]
    
    close=round(df1.iloc[-1]['Close'],2)
    high52=round(df1['Close'].max(),2)
    
    if close >= high52*multiplier:
        diff=close - df1['Close'].max()
        pct_chg=round(abs(diff)/high52*100,2)
        if diff = 0 :
            print("{} is trading at {} ; it's 52W High - {}".format(tick,close,pct_chg,high52))
        elif diff < 0:
            print("{} is trading at {} : {} % below it's 52W High - {}".format(tick,close,pct_chg,high52))
        else:
            print("{} is trading at it's 52W High - {}".format(tick,pct_chg,high52))
```
